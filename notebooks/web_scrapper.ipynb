### 1- First we'll focus on create a functioning code to succesfully retrieve the anime ranking containing the top 50 best-ranked anime

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
r = requests.get("https://myanimelist.net/topanime.php?limit=0")

# Convert to a beautiful soup object
soup = bs(r.content)

table = soup.findChildren('table')[0]
rows = table.findChildren('tr')

# Building the DF
data = []
for row in rows[1:]:
    cols = [col.text.strip() for col in row.find_all('td')]
    data.append(cols)

# Create dataframe
df = pd.DataFrame(data)
df
```

### 2- Replicate the function created instep 1 to now scrapp the complete listv of anime ranking

```python
anime_ranking = pd.DataFrame()
for i in range(100):
    URL_BASE = "https://myanimelist.net/topanime.php?limit="
    rank = i
    url_paginado = URL_BASE + str(rank)
    
    r = requests.get(url_paginado)

    # Convert to a beautiful soup object
    soup = bs(r.content)


    table = soup.findChildren('table')[0]
    rows = table.findChildren('tr')

    data = []
    for row in rows[1:]:
        cols = [col.text.strip() for col in row.find_all('td')]
        data.append(cols)

    # Create dataframe
    df = pd.DataFrame(data)
    anime_ranking=pd.concat([anime_ranking,df])
```
